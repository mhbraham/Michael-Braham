```python
#import data set
import pandas as pd
from google.colab import files
import io

uploaded = files.upload()
if 'Charlotte_zillow_data_2018_2023.csv' in uploaded:
  dfMB = pd.read_csv(io.BytesIO(uploaded['Charlotte_zillow_data_2018_2023.csv']))
else:
  print("File not found. Please make sure to upload the file and try again.")
```

```python
#data investigation
print(dfMB.info())
```

```python
dfMB.describe(include = 'all')
```

```python
# check missing values
dfMB.isnull().sum()/len(dfMB)
```

```python
dfMB['date'] = dfMB['date'].astype('datetime64[ns]')

# printing dataframe
print(dfMB)
print()

print(dfMB.dtypes)
```

```python
dfMB.groupby([(dfMB.date.dt.year),(dfMB.date.dt.month)])['fsinv', 'nlist'].sum()
```

```python
# missing values imputation

NA_means =['nplist', 'avg_dtc', 'med_dtc', 'listpc', 'avg_pcp', 'med_pcd']
dfMB[NA_means]=dfMB[NA_means].fillna(dfMB[NA_means].mean())

dfMB.describe(include='all')
```

```python
# Construct XMB
num_names=['fsinv',	'nlist',	'nplist',	'avg_dpen',	'med_dpen',	'med_list',	'med_sale',	'avg_stlr',	'med_stlr',	'per_alist',	'per_blist',	'avg_dtc',	'med_dtc',	'listpc',
           'avg_pcd',	'avg_pcp',	'med_pcd',	'med_pcp',	'30_yr',	'15_yr' ]

XMB = dfMB[num_names]
XMB.describe(include='all')
```

```python
# skewness in data
XMB.skew(skipna=True)
```

```python
# generate time series data in python
sale_ts = pd.Series(dfMB.med_sale.values,
                     index=dfMB.date,
                     name='sale')

sale_ts
```

```python
sale_ts.describe(include='all')
```

```python
# show visualization

import matplotlib.pylab as plt
ax=sale_ts.plot()
ax.set_xlabel('Time')
ax.set_ylabel('med_sale')
ax.set_ylim(170000,450000)
plt.show()
```

```python
# time series seasonal_decompose
from statsmodels.tsa.seasonal import seasonal_decompose
result=seasonal_decompose(dfMB['med_sale'], model='additive', period=12)
result.plot()
```

```python
# separate into training and validation
nValid = 79
nTrain = len(sale_ts)-nValid

train_ts =sale_ts[:nTrain]
valid_ts = sale_ts[nTrain:]

train_ts
```

```python
import statsmodels.formula.api as sm
from statsmodels.tsa import tsatools, stattools

ts_df = tsatools.add_trend(sale_ts, trend='t')
ts_df['Month'] = ts_df.index.month

ts_df
```

```python
train_df = ts_df[:nTrain]
valid_df = ts_df[nTrain:]

sale_lm = sm.ols(formula='sale~trend',data=train_df).fit()
sale_lm.summary()
```

```python
! pip install dmba
from dmba import regressionSummary

predict_lm = sale_lm.predict(valid_df)
regressionSummary(valid_ts,predict_lm)
```

```python
# quadratic trend
import numpy as np

sale_lm_poly = sm.ols(formula='sale~trend+np.square(trend)',
                       data=train_df).fit()

sale_lm_poly.summary()

predict_lm_poly = sale_lm_poly.predict(valid_df)
regressionSummary(valid_ts,predict_lm_poly)
```

```python
# Seasonality
sale_lm_season = sm.ols(formula='sale~C(Month)',data=train_df).fit()
sale_lm_season.summary()
```

```python
predict_lm_season =sale_lm_season.predict(valid_df)
regressionSummary(valid_ts,predict_lm_season)
```

```python
modelfomula = 'sale~trend+np.square(trend)+C(Month)'
sale_lm_trendseason = sm.ols(formula=modelfomula,
                              data=train_df).fit()

sale_lm_trendseason.summary()
```

```python
predict_lm_trendseason = sale_lm_trendseason.predict(valid_df)
regressionSummary(valid_ts, predict_lm_trendseason)
```

```python
# %matplotlib inline
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plot the actual time series data
plt.plot(sale_ts.index, sale_ts, label='Actual')

# Plot the predicted time series data
plt.plot(predict_lm_trendseason.index, predict_lm_trendseason, label='Predicted')

# Add axis labels and a legend
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()

# Show the plot
plt.show()
```

```python
# smoothing method

ma = train_ts.rolling(12).mean() # create a rolling window of a time series
print(ma)

last_ma = ma[-1]

train_ts
ma

last_ma
predict_ma = pd.Series(last_ma,index=valid_ts.index)
combine_ma = pd.concat([ma, predict_ma])

print(ma)
```

```python
# visualization of moving average
%matplotlib inline
plt.figure(figsize=(20,5))
plt.grid()
plt.plot(train_ts, label='Train')
plt.plot(valid_ts, label='Validation')
plt.plot(combine_ma, label='Simple moving average forecast')
plt.legend(loc='best')
plt.title('Simple Moving Average Method')
plt.show()
```

```python
regressionSummary(valid_ts,predict_ma)
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
SES = SimpleExpSmoothing(train_ts,
                         initialization_method='estimated').fit()
predict_SES = SES.forecast(len(valid_ts))

SES.model.params

regressionSummary(valid_ts, predict_SES)

combine_SES = pd.concat([SES.fittedvalues, predict_SES])
```

```python
%matplotlib inline
plt.figure(figsize=(20,5))
plt.grid()
plt.plot(train_ts, label='Train')
plt.plot(valid_ts, label='Validation')
plt.plot(combine_SES, label='Exponential Smoothing Forecast')
plt.legend(loc='best')
plt.title('Exponential Smoothing Forecast')
plt.show()
```
